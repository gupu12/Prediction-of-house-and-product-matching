```python
amazon = pd.read_csv(r"/amz_com-ecommerce_sample.csv", encoding='unicode_escape')
flipkart = pd.read_csv(r"/flipkart_com-ecommerce_sample.csv", encoding='unicode_escape')
```

```python
ama = amazon[['product_name','retail_price','discounted_price']]
flip = flipkart[['product_name','retail_price','discounted_price']]
```

```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer, util
```

```python
import torch
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Embeddings for amazon and flipkarts products

am_embed = model.encode(amazon['product_name'], convert_to_tensor=True)
Flip_embed = model.encode(flipkart['product_name'], convert_to_tensor=True)
```

```python
## Applying cosine similarity's

cosine_scores = util.cos_sim(am_embed, Flip_embed)
# Taking flipkarts products and comparing with amazon of each product
similar_products = []

for i in range(2000):
    maximum_score = torch.topk(cosine_scores[i], k=5, largest=True, sorted=True).indices
    for ind in maximum_score:
        if ind !=i:
            similar_products.append(ind)
            break
similar_products = [i.item() for i in similar_products]
similar_products
```

```python
flipkart_products = []
amazon_products = []

flipkart_products.append(flip.iloc[similar_products[0:2000]])
amazon_products.append(ama.iloc[0:2000])
```

```python
flipkart_prod_sim = np.reshape(np.array(flipkart_products), newshape=(2000, 3))
flipkart_prod_sim =pd.DataFrame(flipkart_prod_sim)
flipkart_prod_sim.columns = ['Product_Flipkart','Retail_Price_Flipkart','Discounted_Price_Flipkart']
flipkart_prod_sim.head()
```

```python
amazon_products_sim = np.reshape(np.array(amazon_products), newshape=(2000, 3))
amazon_products_sim =pd.DataFrame(amazon_products_sim)
amazon_products_sim.columns = ['Product_name','Retail_Price','Discounted_Price']
amazon_products_sim.head()
```

```python
amazon_products_sim['Retail_Price'] = amazon_products_sim['Retail_Price'].astype(float)
amazon_products_sim['Discounted_Price'] = amazon_products_sim['Discounted_Price'].astype(float)

flipkart_prod_sim['Retail_Price_Flipkart'] = flipkart_prod_sim['Retail_Price_Flipkart'].astype(float)
flipkart_prod_sim['Discounted_Price_Flipkart'] = flipkart_prod_sim['Discounted_Price_Flipkart'].astype(float)

     
```

```python
final_data = pd.concat([flipkart_prod_sim, amazon_products_sim],axis=1)
final_data.to_csv('final_result.csv')
final_data.tail()
```
